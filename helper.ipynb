```python
import matplotlib.pyplot as plt
import pylab
%matplotlib inline
```

```python
x_list = ['NN', 'SVM', 'Fuzzy_Genetic', 'CART', 'Random_Forest']
x = range(len(x_list))
y1_list = [73.6,80.19,69.8,77.45,83.3]
y2_list = [81.1,84.85,83.4,87,86.1]
```

```python
pylab.xticks(x, x_list)
pylab.plot(x, y1_list)
pylab.plot(x, y2_list)
pylab.title("ML Techniques Comparision")
pylab.legend(['Severity','Type'], loc='lower right')
pylab.show()
```
